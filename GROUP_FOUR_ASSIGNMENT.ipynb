```python
# Question 4: Write program to display all prime numbers within a range
start = 1
end = 20

for num in range(start, end + 1):
    if num > 1:
        is_prime = True

        for i in range(2, num):
            if num % i == 0:
                is_prime = False
                break

        if is_prime:
            print(num)
```

```python
# Question 5: Write a program to print a multiplication table of a given number
number = 5   # change this number to any value

for i in range(1, 11):
    print(number, "x", i, "=", number * i)
```
